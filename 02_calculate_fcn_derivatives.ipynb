```python
import torch
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
x_train = torch.arange(0.,2.0*np.pi, 1*np.pi/180.0, requires_grad=True)

print(x_train.shape)
y_train = torch.sin(x_train)
print(y_train.shape)

# loss = torch.ones_like(x_train)

y_train.backward(torch.ones_like(x_train))
# y_train.backward()

print(x_train.shape, y_train.shape)

```

```python
plt.figure(figsize=(8,6))

plt.plot(x_train.detach().numpy(), y_train.detach().numpy(), label="Orig")
plt.plot(x_train.detach().numpy(), x_train.grad.detach().numpy(), label="Gradient")

plt.legend()
plt.grid(True)
```

```python

# Define input range
x = torch.linspace(-math.pi, math.pi, 200, requires_grad=True)

print(x.shape)
# Calculate sin(x)
y = torch.sin(x)

# Since we're not calculating any loss, provide a dummy gradient
y.backward(torch.ones_like(x))

# Plot sin(x) and its gradient (cos(x))
plt.plot(x.detach().numpy(), y.detach().numpy(), label='sin(x)')
plt.plot(x.detach().numpy(), x.grad.detach().numpy(), label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
